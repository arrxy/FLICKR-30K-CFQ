## deal_raw_enhanced

```python
import json
import re
import openai
from tqdm import tqdm
import random
from config import PROMPT_LIST
```

```python
openai.api_key = "EMPTY" # Not support yet
openai.api_base = "http://localhost:8030/v1"
# openai.api_base = "http://localhost:16363/v1"
model = "vicuna-13b-v1.1"
```

```python
text_to_raw_enchanced = json.load(open('data/text_to_raw_enhanced.json', 'r', encoding='utf-8'))
```

```python
new_text_to_enhanced = {}
for text, raw_enhanced in text_to_raw_enchanced.items():
    r = re.sub('(\d.)', ' ', r)
    enhance_list = r.strip().split('\n')
    new_text_to_enhanced[text] = []
    if len(enhance_list) > 1:
        for enhance in enhance_list:
            if len(enhance) > 0:
                if enhance[0].isalpha() == False:
                    new_text_to_enhanced[text].append(enhance[1:].strip())
                else:
                    new_text_to_enhanced[text].append(enhance.strip())
```

```python
json.dump(new_text_to_enhanced, open('data/text_to_enhanced_list.json', 'w', encoding='utf-8'), ensure_ascii=False)
```

## Double check

二次确认enhance后的文本是否是噪音

```python
import json
from config import CHECK_PROMPT_LIST
import openai
from tqdm import tqdm
import random
```

```python
text_to_enhanced_list = json.load(open('data/text_to_enhanced_list.json', 'r', encoding='utf-8'))
```

```python
text_to_raw_checked= {}
prompt = CHECK_PROMPT_LIST[0]
```

```python
for text, enhance_list in text_to_enhanced_list.items():
    for enhance_text in enhance_list:
        content = prompt.format(enhance=enhance_text, text=text)
        completion = openai.ChatCompletion.create(
          model=model,
          messages=[{
             "role": "user",
             "content": content
          }]
        )
        # print the completion
        result = completion.choices[0].message.content
        text_to_raw_checked[text] = result
json.dump(text_to_raw_checked, open('data/text_to_raw_checked.json', 'w', encoding='utf-8'))
```
