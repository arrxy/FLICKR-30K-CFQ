## 检查flickr30k test的图片是否全不在flickr30k entities中

```python
import os
import shutil
```

```python
img_dir_flickr30k_entities = '/Users/aidan/Learn/NLP/Retrieval/Dataset/Flickr30k_Entities/{}'
```

```python
img_files_flickr30k_test = []
with open('data/flickr30k_test.txt', 'r', encoding='utf-8') as f:
    for line in f:
        img_files_flickr30k_test.append(line.strip())
```

```python
for img_file in img_files_flickr30k_test:
    to_path = os.path.join('img', img_file)
    flag = 0
    for type in ['train', 'val', 'test']:
        img_dir = img_dir_flickr30k_entities.format(type)
        from_path = os.path.join(img_dir, img_file)
        if os.path.exists(from_path):
            shutil.copy(from_path, to_path)
            flag = 1
    if flag == 0:
        print(img_file)
```
