```python
import json
import os
import sys
sys.path.append(main_dir_path)

from tools import  *
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
tag_map_file_path = 'data/img_to_tags.json'
img_dir_path = os.path.join(main_dir_path, 'images')
```

```python
img_to_tags = read_json(tag_map_file_path)
```

## 显示一张图片

```python
img_file_name = random.choice(list(img_to_tags.keys()))
img_file_name, img_to_tags[img_file_name]
```

```python
img_file_path = os.path.join(img_dir_path, '{}.jpg'.format(img_file_name))
img = Image.open(img_file_path)

plt.figure("Image") # 图像窗口名称
plt.imshow(img)
plt.axis('off') # 关掉坐标轴为 off
plt.title('\n'.join(img_to_tags[img_file_name])) # 图像题目
plt.show()
```

## 显示多张图片

```python
row_num = 2 # 行数
col_num = 4 # 列数
sum_num = row_num * col_num

img_file_name_list = random.sample(list(img_to_tags.keys()), sum_num)
[(img_file_name, img_to_tags[img_file_name]) for img_file_name in img_file_name_list ]
```

```python

plt.figure(figsize=(10,5)) #设置窗口大小
for row_i in range(row_num):
    for col_i in range(col_num):
        img_file_name = img_file_name_list[row_i * col_num + col_i]
        img_file_path = os.path.join(img_dir_path, '{}.jpg'.format(img_file_name))
        img = Image.open(img_file_path)

        tags = img_to_tags[img_file_name]

        plt.subplot(row_num, col_num, row_i * col_num + col_i + 1), plt.title('\n'.join(tags))
        plt.imshow(img), plt.axis('off')
```
